```python
#data loading
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
bat = pd.read_csv(r"bat.csv")
match = pd.read_csv(r"match.csv")
```

```python
bat['batting_score'] = (
    bat['runs_x'] * 0.4 +
    bat['sixes'] * 0.3 +
    bat['fours'] * 0.2 +
    bat['strike_rate'] * 0.1
)
```

```python
# Merge
df = pd.merge(
    bat,
    match[['match_number', 'scorecard_id',
           'matchtype', 'start_date',
           'runs', 'over', 'run_rate',
           'opp_team_id', 'team_id']],
    on=['match_number', 'scorecard_id'],
    how='left'
)

print(df)
```

```python
df.to_csv("merged_dataset.csv", index=False)
```

```python
df = df.drop(columns=[
    'matchtype_y', 'start_date_y',
    'runs_y', 'over_y', 'run_rate_y',
    'opp_team_id_y', 'team_id_y', 'runs'
])
```

```python
df = df.drop('fall_of_wicket', axis=1)
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True, errors="ignore")
# Check Missing Values
print(df.isnull().sum())
```

```python
# Drop rows with ANY null values
df.dropna(inplace=True)

# Drop rows where a SPECIFIC column is null
df.dropna(subset=['playing_role','match_number','name_y','matchtype_x'], inplace=True)

# Drop rows where ALL values are null
df.dropna(how='all', inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Finding Duplicates
df.duplicated().sum()
```

```python
# Convert categorical columns to string to avoid errors during encoding
df['name_x'] = df['name_x'].astype(str)
df['name_y'] = df['name_y'].astype(str)

# Remove extreme outliers (e.g., scores > 200) to improve model stability
df = df[df['runs_x'] < 200]
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder

le_player = LabelEncoder()
le_team = LabelEncoder()

# Encode player names and team names
df['player_encoded'] = le_player.fit_transform(df['name_x'])
df['team_encoded'] = le_team.fit_transform(df['name_y'])
```

```python
#feature selection
# X = Input Features, Y = Target Variable
# Standard features include balls faced, the over they played in, the player, and the opposition
X = df[['balls', 'over_x', 'player_encoded', 'team_encoded']]
Y = np.log1p(df['runs_x'])  # Using log transform helps normalize skewed run data
```

```python
# LEAST GRADIENT BOOSTING SYSTEM (LGBM)
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Log transform target
Y = np.log1p(df['runs_x'])

# Feature Scaling
scaler = StandardScaler()
X[['balls','over_x']] = scaler.fit_transform(X[['balls','over_x']])

# Train Test Split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

# LightGBM Model (Tuned)
model = LGBMRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=8,
    num_leaves=31,
    random_state=42
)

model.fit(X_train, Y_train)

# Prediction
Y_pred = model.predict(X_test)

# Convert back from log scale
Y_pred = np.expm1(Y_pred)
Y_test = np.expm1(Y_test)


# Model Evaluation
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R² Score:", r2)


# Player Prediction
player = input("Enter the player name: ")
opposition = input("Enter the 'v India': ")

player_data = df[(df['name_x'] == player) & (df['name_y'] == opposition)]

if not player_data.empty:

    X_player = player_data[['balls','over_x','player_encoded','team_encoded']]
    X_player[['balls','over_x']] = scaler.transform(X_player[['balls','over_x']])

    predicted_runs = model.predict(X_player)

    # Convert back to runs
    predicted_runs = np.expm1(predicted_runs)

    predicted_runs_int = np.round(predicted_runs).astype(int)

    print(f"\nPredicted runs for {player} against {opposition}: {predicted_runs_int}")
    print(f"Latest match prediction: {predicted_runs_int[-1]}")
    print(f"Average predicted runs: {predicted_runs_int.mean()}")

else:
    print("No data available for the specified player and opposition.")
```

```python
import pickle
pickle.dump({
    'model': model,
    'scaler': scaler,
    'player_encoder': le_player,
    'team_encoder': le_team
}, open('cricket_lgbm_model.pkl', 'wb'))

print("Pickle file created: cricket_lgbm_model.pkl")
```
